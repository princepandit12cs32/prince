```python
                                            
                                                # Python Pandas
    

import pandas as pd

# Single Dimensional


n1 = pd.Series([1,22,12,32,2.1])

#print(n1)
#print(type(n1))

# Changing the index of table

n2 = pd.Series([12,56,'Ram',33,11,21], index = ['a','b','c','d','e','f'])
n2
```

```python
# Series Objects from Dictionary

p1 = pd.Series({'a':11,'b':22,'c':13.4})

#print(p1)




# Checking Index position

p2 = pd.Series({'a':11,'b':22,'c':13.4}, index = ['b','c','d','a'])
p2



```

```python
# Extracting individual Elements

n1 = pd.Series([11,45,34,'Ram','Mango',35])
print(n1[3]) # Extract a single element



print(n1[:4]) # Extracting sequence of elements


print(n1[-2:]) # Extracting element from back of series
```

```python
                                    # Basic Math Operations on Series
# Adding and substracting a Scalar value to series
n1 = pd.Series([122,313,12312,23124,11])

n1 = n1 + 100 # Addition
#print(n1)

n2 = n1 - 100 # Substraction

#print(n2)



# Adding two Series objects

s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([1,2,3,4,5])

print(s1+s2) # addition

print(s1-s2) # Substraction
```

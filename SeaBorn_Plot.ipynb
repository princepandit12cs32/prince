```python
#   SeaBorn Line Plot

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

fmri = pd.read_excel("products.xlsx")
fmri.head

sns.lineplot(x="name",y = "price",data=fmri)
plt.grid(True)
plt.show()

# Grouping data with 'hue'

sns.lineplot(x='month',y='price',data=fmri,hue="price")
plt.show() 
```

```python
# Adding Style

fmri = pd.read_excel("iris.xlsx")

sns.lineplot(x="Customer",y="Qtr 1",data=fmri,hue="Product",style="Product")
plt.show()
```

```python
#    Adding Markers


sns.lineplot(x="Customer",y="Qtr 1",data=fmri,hue="Product",style="Product",markers=True)
plt.show()
```

```python
#              SeaBorn BAR - PLOT


import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns


data = pd.read_excel("players.xlsx")
data.head()

sns.barplot(x="Team", y="Weight", data=data)
sns.set(style = "whitegrid")
plt.show()
```

```python
# Grouping Data With "hue"

sns.barplot(x="Country", y="Weight", hue="Team",data=data)
plt.title("Bar Plot of Hocky Players")
#sns.barplot(x="Team", y="Weight", data=data)
plt.show()
```

```python
# Adding stlye:


sns.barplot(x="Team", y="Weight", data=data, palette="Blues_d",hue="Country")

plt.show()

# Adding stlye(color):


sns.barplot(x="Team", y="Weight", data=data,hue="Country",color="orange")

plt.show()
```

```python
#  SeaBorn Scatter Plot

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

players = pd.read_excel("players.xlsx")

players.head()

sns.scatterplot(x="Team", y="Age", hue = "Country", data = players)
plt.show()
```

```python
#         SeaBorn     Histogram 

sns.distplot(players["Age"],color="red") # distplot() : function is used for show the histogram and line graph and take argument numeric
plt.show()
```

```python
sns.displot(players["Age"],color="red") # displot show only histogram not line graph
plt.show()
```

```python
#         SeaBorn     Histogram 

sns.distplot(players["Weight"],hist=False,color="red") # removing hist
plt.show()
```

```python
# adding style:

#         SeaBorn     Histogram 

sns.distplot(players["Age"],bins=50,kde=False) # kde is remove the lime graph and bins is show the width of histogram
plt.show()
```

```python
#    Vertical     SeaBorn     Histogram 

sns.distplot(players["Age"],vertical=True) # distplot() : function is used for show the numeric values
plt.show()
```

```python

#       Seaborn JointPlot

sns.jointplot(x="Team",y="Age",data=players,color="orange")
plt.show()
```

```python
# joint plot seprate by line

sns.jointplot(x="Team",y="Age",data=players,kind="reg",color="red")
plt.show()
```

```python
##   SeaBorn BoxPlot



#         SeaBorn     Histogram 

sns.boxplot(x="Team",y="Age",data=players,hue="Country")

plt.show()
```

```python
#   07.  SeaBorn PairPlot

sns.pairplot(players,hue="Team")
plt.show()
```

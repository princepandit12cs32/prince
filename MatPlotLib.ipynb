```python
import numpy as np
from matplotlib import pyplot as plt

#     ------------------------ 01 Line Plot ----------------------------

x = np.arange(1,11)
y = x*0.4

plt.plot(x,y) # assigning x and y points
plt.title("Line Plot") # Display the TITLE
plt.xlabel("X - Axis") # Display the X axis title
plt.ylabel("Y - Axis") # Display the Y axis title
plt.show()


# ---------------------------------------Changing Line Aesthetics

plt.plot(x,y,color='r',linestyle=":",linewidth="5")
plt.title("Line Plot Display with in Grid") # Display the TITLE
plt.xlabel("X - Axis") # Display the X axis title
plt.ylabel("Y - Axis") # Display the Y axis title
plt.grid(True) # Display background grid
plt.show()


#------------------------------- Adding two Lines in the same plot
y1 = x*9

plt.plot(x,y,color='r',linestyle=":",linewidth="5")
plt.plot(x,y1,color="g",linewidth="5")
plt.title("Adding two Lines in the same plot") # Display the TITLE
plt.xlabel("X - Axis") # Display the X axis title
plt.ylabel("Y - Axis") # Display the Y axis title
plt.grid(True) # Display background grid
plt.show()

```

```python
# Adding Sub-Plot (Vertical)

x1 = np.arange(10,101)

y1 = x1*2
y2 = x1*.5612
x2 = x1*12.12

plt.subplot(1,2,1)
plt.plot(x,y1,color="r",linewidth="5")
plt.title("Sub-Plot 01")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x2,y2,color="y",linewidth="5")
plt.title("Sub-Plot 02")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
plt.grid(True)
plt.show()
```

```python
# Adding two sub-plot in same plot(Horizontal)

x1 = np.arange(1.3,50)
y1 = x1*22.03

x2 = np.arange(12,112)
y2 = x2*2

plt.subplot(2,1,1)
plt.plot(x1,y1, color="g",linewidth="5")
plt.title("Sub Plot 01")
#plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x2,y2, color="orange",linewidth="5")
plt.title("Sub Plot 02")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid(True)
plt.show()
```

```python
#     -------------------------  02 Bar Plot  ---------------------------------------

students = {"Bob":341,"Sam":475,"Matt":230,"Prince":500}

names = list(students.keys())
marks = list(students.values())

plt.subplot(1,2,1)
plt.bar(names,marks) # Vetical View
plt.title("Students Bar Plot")
plt.xlabel("<------------ Names -------------->")
plt.ylabel("<------------ Marks -------------->")
plt.grid(True)

plt.subplot(1,2,2)
plt.barh(names,marks,color="red") # Horizontal View

plt.show()
```

```python
#    --------------------    03 Scatter Plot  -----------------------

x = np.arange(1,11)
a = (12,3,4,22,33,1,11,9,10,12)
b = (1.2,3.4,2.3,11.2,.5,6.6,7.6,8,9,10)

plt.scatter(x,a, color="red", marker="*",s = 100)
plt.scatter(x,b, marker="^",s= 100) # Adding two scatter in same graph
plt.grid()
plt.show()



# Adding two scatter in sub plot

plt.subplot(1,2,1)
plt.scatter(x,a,color="r",marker = "*", s = 100)
plt.subplot(1,2,2)
plt.scatter(x,b,color="green",marker = "^", s = 100)
plt.show()
```

```python
#      06.    Violin - Plot

from matplotlib import pyplot as plt

d1 = [12,23,34,45,56,67]
d2 = [1.2,12.3,3.4,4.5,15.6]
d3 = [11,22,33,44,55]

data = list([d1,d2,d3])

plt.violinplot(data, showmedians=True)
plt.grid(True)
plt.show()
```

```python
#     07.      Pie-Chart

from matplotlib import pyplot as plt

# creating data
fruits = ["Orange","Guava","Mango","Kiwi","Apple"]
quantity = [12,13,23.4,32,44]

# Making pie-chart plot

plt.pie(quantity,labels=fruits)
plt.show()

# Changing Aesthetics

plt.pie(quantity,labels=fruits,autopct="%0.1f%%",colors=['orange','yellow','green','blue','red'])
plt.title("Fruits Pie Chart")
plt.show()
```

```python
#      08.  DoughNut - Chart

# creating data

animals = ['Cat','Rat','Oxe',"Fox","Horse"]
quantity = [12,34,23,43,8]

# creating plot

plt.pie(quantity,labels=animals,radius=1,autopct="%0.1f%%")
plt.pie([1],colors=['white'],radius=.4)
plt.title("Animal Chart")
plt.show()
```

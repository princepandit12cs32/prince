```python
# making bar plot for the top 10 players age
plt.figure(figsize=(10,5))
plt.bar(list(pls["NameF"].value_counts()[0:10].keys()),list(pls["Age"].value_counts()[0:10]))
plt.show()
```

```python
# making bar plot for the top 20 hometown and weight

plt.figure(figsize=(25,10))
plt.bar(list(pls["Hometown"].value_counts()[0:20].keys()),list(pls["Weight"].value_counts()[0:20]),color="r")
plt.title("Players Home Town and Weight")
plt.xlabel("<--        Hometown          -->")
plt.ylabel("<--                    Weight                 -->")
plt.show()
```

```python
#list the top 20 Height of the players

list(pls["Height"].value_counts()[0:20])
```

```python
# list the top 20 home town of the players

list(pls["Hometown"].value_counts()[0:20].keys())
```

```python
# Pie Chart of top 10 hoetown and BMI of players
plt.figure(figsize=(7,5))
plt.pie(list(pls["BMI"].value_counts()),labels=list(pls["Hometown"].value_counts()[0:10].keys()),radius=2,autopct="%.1f%%")
plt.pie([1],colors="w",radius=1)
plt.show()
```
